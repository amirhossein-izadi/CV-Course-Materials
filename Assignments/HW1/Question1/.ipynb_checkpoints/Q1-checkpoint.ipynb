```python
import cv2 as cv
from PIL import Image
import matplotlib.pyplot as plt 
import numpy as np
```

### load Image 

```python
image_path = 'Sample.jpeg'
image = cv.imread(image_path, cv.IMREAD_GRAYSCALE)
```

## part A

```python
hist = cv.calcHist([image], [0], None, [256], [0, 256])

plt.plot(hist)
plt.title('Image Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

## Part B

```python
def histogram_slide(image, direction, factor):
    
    if direction == 'dark':
        result = image - factor
    elif direction == 'light':
        result = image + factor
    else:
        raise ValueError("Invalid direction. Use 'dark' or 'light'.")

    image = np.clip(result, 0, 255).astype(np.uint8)

    return image


dark_image = histogram_slide(image, 'dark', 50)
light_image = histogram_slide(image, 'light', 50)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1), plt.imshow(light_image, cmap='gray'), plt.title('Light Image')
plt.subplot(1, 3, 2), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(1, 3, 3), plt.imshow(dark_image, cmap='gray'), plt.title('Dark Image')
plt.show()

```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1), plt.hist(light_image.ravel(), 256, [0, 256], color='lightblue', alpha=0.7)
plt.title('Histogram - Light Image')

plt.subplot(1, 3, 2), plt.hist(image.ravel(), 256, [0, 256], color='gray', alpha=0.7)
plt.title('Histogram - Original Image')

plt.subplot(1, 3, 3), plt.hist(dark_image.ravel(), 256, [0, 256], color='darkblue', alpha=0.7)
plt.title('Histogram - Dark Image')

plt.show()
```

## Part C

```python
def histogram_equalization_manual(image):

    hist, bins = np.histogram(image.flatten(), 256, [0, 256])
    cdf = hist.cumsum()
    cdf_normalized = cdf / float(cdf.max())

    equalized_image = np.interp(image.flatten(), bins[:-1], cdf_normalized) * 255
    equalized_image = equalized_image.reshape(image.shape).astype(np.uint8)

    return equalized_image
```

```python
equalized_image_manual = histogram_equalization_manual(image)
equalized_image_opencv = cv.equalizeHist(image)


plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(1, 3, 2), plt.imshow(equalized_image_manual, cmap='gray'), plt.title('Equalized (Manual)')
plt.subplot(1, 3, 3), plt.imshow(equalized_image_opencv, cmap='gray'), plt.title('Equalized (OpenCV)')
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1), plt.hist(image.ravel(), 256, [0, 256], color='gray', alpha=0.7)
plt.title('Histogram - Original Image')

plt.subplot(1, 3, 2), plt.hist(equalized_image_manual.ravel(), 256, [0, 256], color='lightgreen', alpha=0.7)
plt.title('Histogram - Manually Equalized Image')

plt.subplot(1, 3, 3), plt.hist(equalized_image_opencv.ravel(), 256, [0, 256], color='lightcoral', alpha=0.7)
plt.title('Histogram - OpenCV Equalized Image')

plt.show()
```
